```python
# how to optimize linear separator
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn import svm
```

```python
data = pd.read_csv('files/gender.csv')
data.head() # centimeters
```

```python
data['Class'] = data['Gender'].apply(lambda x: 'r' if x == 'F' else 'b')
data.head()
```

```python
data = data.iloc[:25] # minimize our dataset
```

```python
# demonstrating that a separator could be very different
fig, ax = plt.subplots()
ax.scatter(x=data['Height'], y=data['Hair length'], c=data['Class'])
ax.axhline(20) # sepation line1
x = np.array([150, 190])
ax.plot(x, x - 145, c ='y') # sepation line2
ax.plot(x, 2*x - 310, c='g') # sepation line3

# all these lines are separating but which one is better? 
```

```python
X = data[['Height', 'Hair length']]
y = data['Gender']
y = np.array([0 if gender == 'M' else 1 for gender in y])
```

```python
clf = svm.SVC(kernel='linear')
clf.fit(X, y)
```

```python
X_test = np.random.rand(10000, 2)
X_test = X_test*(50, 120) + (150, 0)

y_pred = clf.predict(X_test)
```

```python
fig, ax = plt.subplots()
ax.scatter(x=X_test[:, 0], y=X_test[:, 1], c=y_pred, alpha=.25)
# line of the separation of these points : 

# transform our color (    )
y_color = ['r' if value == 0 else 'b' for value in y]
ax.scatter(x=X['Height'], y=X['Hair length'], c=y_color)
# this is the best line to separate the maximum between the values
```
